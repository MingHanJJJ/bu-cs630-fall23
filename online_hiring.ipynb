```python
import matplotlib.pyplot as plt
import numpy as np

n = 1000
k_values = np.arange(1, 1001)
f_values = [k/n * np.log(n/k) for k in k_values]

plt.figure(figsize=(10, 5))
plt.plot(k_values, f_values, label=r'$f(k) = \frac{k}{n} \ln(\frac{n}{k})$')
plt.xlabel('k')
plt.ylabel('f(k)')

# Drawing the vertical line at x=1000/e
plt.axvline(x=n/np.e, color='r', linestyle='--', label=r'$k=\frac{n}{e}$')

plt.legend()
plt.title(r'$f(k)$ and vertical line at $xk=\frac{n}{e}$')
plt.grid(True)
plt.show()
```
