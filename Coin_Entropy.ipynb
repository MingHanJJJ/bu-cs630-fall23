```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def entropy(p):
    if p == 0 or p == 1:
        return 0
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)

p_values = np.linspace(0, 1, 1000)
h_values = [entropy(p) for p in p_values]
plt.plot(p_values, h_values, '-b')
plt.title('Entropy of a Coin Toss as a Function of Probability of Heads')
plt.xlabel('Probability of Heads (p)')
plt.ylabel('Entropy (H)')
plt.axvline(x=0.5, color='red', linestyle='--', label='p=0.5')
plt.grid()
plt.show()
```
